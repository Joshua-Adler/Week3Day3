### Exercise #1 <br>
<p>Reverse the list below in-place using an in-place algorithm.<br>For extra credit: Reverse the strings at the same time.</p>

```python
words = ['this' , 'is', 'a', 'sentence', '.']

#Question: Isn't it impossible to reverse a string in-place because they're immutable?
def reverserino(thing):
    for i in range(len(thing) // 2):
        other = len(thing) - i - 1
        #String reversing is not in-place
        thing[i], thing[other] = thing[other][::-1], thing[i][::-1]
        
reverserino(words)
print(words)
```

### Exercise #2 <br>
<p>Create a function that counts how many distinct words are in the string below, then outputs a dictionary with the words as the key and the value as the amount of times that word appears in the string.<br>Should output:<br>{'a': 5,<br>
 'abstract': 1,<br>
 'an': 3,<br>
 'array': 2, ... etc...</p>

```python
import re
a_text = 'In computing, a hash table hash map is a data structure which implements an associative array abstract data type, a structure that can map keys to values. A hash table uses a hash function to compute an index into an array of buckets or slots from which the desired value can be found'

def count_words(string):
    #Get all the words
    words = re.compile(r'\w+').findall(string.lower())
    #Sorting not necessary, only done to match example output
    words.sort()
    #Count all the words
    dist = {}
    for word in words:
        if word in dist:
            dist[word] += 1
        else:
            dist[word] = 1
    return dist

print(count_words(a_text))
```

## Exercise #3

Write a program to implement a Linear Search Algorithm. Also in a comment, write the Time Complexity of the following algorithm.

#### Hint: Linear Searching will require searching a list for a given number. 

```python
import random

#O(n)
#Returns None if number not found
#Probably not necessary, but muh readability
def search(nums, targ):
    for i in range(len(nums)):
        if nums[i] == targ:
            return i
    return None

#Make a random list with all numbers 1-10
nums = [i + 1 for i in range(10)]
random.shuffle(nums)
print(f'List: {nums}')

#Conduct some searches
for i in range(1, len(nums) + 1):
    print(f'{i} is at index {search(nums, i)}')
```
